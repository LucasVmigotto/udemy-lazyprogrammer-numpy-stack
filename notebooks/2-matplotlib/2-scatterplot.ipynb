## Scatter plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.random.randn(100, 2)

X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
X = np.random.randn(200, 2)

X[:50] += 3

X.shape
```

```python
Y = np.zeros(200)

Y[:50] = 1

Y.shape
```

```python
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=Y)
```
