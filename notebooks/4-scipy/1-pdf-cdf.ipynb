# PDF and CDF

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x = np.linspace(-6, 6, 1000)

x.shape
```

```python
fx = norm.pdf(x, 
              loc=0, # Mean
              scale=1) # Standard deviation

fx.shape
```

```python
plt.plot(x, fx)
```

```python
Fx = norm.cdf(x, loc=0, scale=1)

Fx.shape
```

```python
plt.plot(x, Fx)
```

```python
log_fx = norm.logpdf(x, loc=0, scale=1)

log_fx.shape
```

```python
plt.plot(x, log_fx)
```

```python
log_Fx = norm.logcdf(x, loc=0, scale=1)

log_Fx.shape
```

```python
plt.plot(x, log_Fx)
```
