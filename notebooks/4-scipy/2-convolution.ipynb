# Convolution

```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
import requests

LENA_URL = 'https://upload.wikimedia.org/wikipedia/en/thumb/7/7d/Lenna_%28test_image%29.png/220px-Lenna_%28test_image%29.png'
LENA_PATH = './data/raw/lena.png'

with open(LENA_PATH, 'wb') as f:
    f.write(requests.get(LENA_URL).content)

lena = Image.open(LENA_PATH)

type(lena)
```

```python
gray = np.mean(lena, axis=2)
```

```python
x = np.linspace(-6 , 6, 50)
```

```python
from scipy.stats import norm

fx = norm.pdf(x, loc=0, scale=1)
```

```python
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='grey')
```

```python
from scipy.signal import convolve2d

out = convolve2d(gray, filt)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray')
```
